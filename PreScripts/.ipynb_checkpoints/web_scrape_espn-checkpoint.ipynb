```python
# import beautiful soup
from bs4 import BeautifulSoup as bs

# import Pandas
import pandas as pd

# import splinter for automated browser manipulation
from splinter import Browser

# import request library for http requests
import requests

# import time
import time
```

```python
# ESPN url to be scraped
url = "http://www.espn.com/nba/salaries"

# module to call API
response = requests.get(url)

# scrape raw text from page
soup = bs(response.text, "html.parser")

# print soup
print(soup.prettify())
```

```python
# generate an empty table to store all the info
col_names = [0, 1, 2, 3, 4] # 4 is one extra for the year
all_table = pd.DataFrame(columns = col_names)

# A loop to navigate through the years
for year in range(2019, 1999, -1):
    # A loop to click through the pages within a year
    page = 0
    while(True):
        page += 1
        # scrape table from page
        # read table into pandas
        current_url = url + "/_/year/" + str(year) + "/page/" + str(page)
        tables = pd.read_html(current_url)
        table = tables[0]
        # if we get an empty table, break
        if len(table) <= 1:
            break
        else:
            # add the column standing for the year to the table
            table[4] = year
            # append the new table to table
            print(f"Now storing the content of page {page} of year {year}!")
            all_table = pd.concat([all_table, table])
```

```python
# Rename the columns
all_table_out = all_table.rename(columns={0: "Rank in the year", 1: "Name", 2: "Team", 3: "Salary", 4: "Year"})

# Get rid of the title lines given by the website
all_table_out = all_table_out[all_table_out["Rank in the year"] != 'RK']

# re-index
all_table_out = all_table_out.reset_index()
#all_table_test = all_table_test.drop("index")
all_table_out = all_table_out[["Rank in the year", "Name", "Team", "Salary", "Year"]]
all_table_out.head()
```
