```python
# import Pandas
import pandas as pd

# import os
import os

# import numpy
import numpy as np

# import pyplot
import matplotlib.pyplot as plt
```

```python
# Input data
# 1) NBA Player Stats since 1950
# https://www.kaggle.com/drgilermo/nba-players-stats
# Flat file (CSV) dataset from Kaggle
player_stats_inp = pd.read_csv(os.path.join("..", "RawData", "Seasons_Stats.csv"))

# 2) NBA Play Salaries 1999-2019
# http://www.espn.com/nba/salaries
# Structured dataset scrapped from web
# See PreScripts/web_scrape_espn.ipynb for the scraping process
player_salary_inp = pd.read_csv(os.path.join("..", "RawData", "player_salary.csv"))

# 3) NBA Salary Cap by History
# https://www.basketball-reference.com/contracts/salary-cap-history.html
# SQL dataset (originally CSV, exported to SQL externally using Python Alchemy)
salary_cap_inp = pd.read_csv(os.path.join("..", "CleanData", "salary_cap_clean.csv"))
```

```python
# Clean up dataset 1 to get ready for merging
# keep useful columns
player_stats = player_stats_inp[["Year", "Player", "Age", "FG%", "eFG%", "FT%", "PTS"]]

# Get rid of NaN values
player_stats = player_stats.dropna()

# Make the values that should be integer integer
player_stats.Year = player_stats.Year.astype(int)
player_stats.Age = player_stats.Age.astype(int)

# rname column
player_stats = player_stats.rename(columns={"Player": "Name"})

player_stats.head()
```

```python
# Clean up dataset 2 to get ready for merging
player_salary = player_salary_inp[["Name", "Position", "Team", "Salary", "Year"]]


player_salary.tail()
```

```python
# Clean up salary cap data
salary_cap = salary_cap_inp.rename(columns={"SalaryCap": "Salary Cap"})
salary_cap["Salary Cap"] = salary_cap["Salary Cap"].astype(int)

salary_2018 = salary_cap[salary_cap["Year"] == 2018]["Salary Cap"]

# Get the value of most recent year
salary_2018 = salary_2018.values[0]
```

```python
# Inner merge datasets 1 and 2
player_stats_salary = player_stats.merge(player_salary, how = "inner", on = ["Name", "Year"])
player_stats_salary.head()
```

```python
# Convert Salary to numbers
player_stats_salary["Salary"] = player_stats_salary["Salary"].str.strip("$")
player_stats_salary["Salary"] = player_stats_salary["Salary"].str.replace(',', '').astype(int)
player_stats_salary.head()
```

```python
# Normalize with salary cap by the year
player_normalized = player_stats_salary.merge(salary_cap, how = "inner", on = "Year")

player_normalized["Salary Normalized"] = player_normalized["Salary"] / player_normalized["Salary Cap"] * salary_2018

player_normalized.tail()
```

```python
# analyze position info
# The five basketball positions normally employed by organized basketball teams are 
# the point guard (PG), the shooting guard (SG), the small forward (SF), the power forward (PF), and the center (C).
player_gby = player_normalized.groupby("Position")
player_gby.describe()
```

```python
# ditch wrong data
player_normalized["Position"] = player_normalized["Position"].str.strip()
player_normalized = player_normalized[player_normalized["Position"] != "GF"]
player_normalized.head()

# Change all ambivalent positions to its less valuable possibility because we don't have time for their feelings
# Basically, all Fs will be Power Forwards, and all Gs will be Shooting Guards
# Sorry not sorry
player_normalized["Position"] = player_normalized["Position"].replace("F", "PF").replace("G", "SG")
player_normalized.head()
```

```python
# analyze position info again
# The five basketball positions normally employed by organized basketball teams are 
# the point guard (PG), the shooting guard (SG), the small forward (SF), the power forward (PF), and the center (C).
player_gby = player_normalized.groupby("Position")
player_gby.describe()
```

```python
player_normalized['Salary Normalized'] = player_normalized['Salary Normalized']*0.000001
player_normalized.rename(index=str, columns={"Salary Normalized": "Salary Normalized (Million)"}, inplace=True)
player_normalized.head()
```

```python
# First plot: salary vs. position
player_gby_mean_df = player_gby.mean()
player_gby_mean_df = player_gby_mean_df.reindex(['PG','SG','SF','PF','C'])
# player_gby_mean_df['Salary Normalized'] = player_gby_mean_df['Salary Normalized']*0.000001
# player_gby_mean_df.rename(index=str, columns={"Salary Normalized": "Salary Normalized (Million)"}, inplace=True)
player_gby_mean_df
```

```python
x_axis = np.arange(len(player_gby_mean_df))
tick_locations = [value for value in x_axis]
plt.figure(figsize=(10,6))
plt.bar(x_axis, player_gby_mean_df['Salary Normalized (Million)'], color='steelblue', alpha=1, align="center")
plt.xticks(tick_locations, list(player_gby_mean_df.index))
plt.xlabel("Position")
plt.ylabel("Average Salary (Million) Normalized to 2018")
plt.title("Salary versus Position")
plt.savefig("../Plots/Salary versus Position.jpg")
```

```python
player_gby_age = player_normalized.groupby("Age")
player_gby_age.describe()
player_mean_age_df = player_gby_age.mean()
player_mean_age_df
```

```python
x = list(player_mean_age_df.index)
y = player_mean_age_df["Salary Normalized (Million)"]
color = ["steelblue"]
plt.xlabel('Age')
plt.ylabel('Average Salary (Million) Normalized to 2018')
plt.title("Salary versus Age")
plt.scatter(x, y, marker=".", alpha=1, c=color, s=100)
plt.savefig("../Plots/Salary versus Age.jpg")
plt.show()
```

```python
player_normalized['FG%_bin'] = player_normalized['FG%']//0.05*0.05
player_normalized.head()
player_gby_fg = player_normalized.groupby("FG%_bin")
player_gby_fg.describe()
player_mean_fg_df = player_gby_fg.mean()
player_mean_fg_df
```

```python
x = player_mean_fg_df.index
y = player_mean_fg_df["Salary Normalized (Million)"]
color = ["steelblue"]
plt.xlabel('Field Goal Percentage')
plt.ylabel('Average Salary (Million) Normalized to 2018')
plt.title("Salary versus FG%")
plt.scatter(x, y, marker=".", alpha=1, c=color, s=100)
plt.savefig("../Plots/Salary versus FG%.jpg")
plt.show()
```

```python
player_normalized['FT%_bin'] = player_normalized['FT%']//0.05*0.05
player_normalized.head()
player_gby_ft = player_normalized.groupby("FT%_bin")
player_gby_ft.describe()
player_mean_ft_df = player_gby_ft.mean()
player_mean_ft_df
```

```python
x = player_mean_ft_df.index
y = player_mean_ft_df["Salary Normalized (Million)"]
color = ["steelblue"]
plt.xlabel('Free Throw Percentage')
plt.ylabel('Average Salary (Million) Normalized to 2018')
plt.title("Salary versus FT%")
plt.scatter(x, y, marker=".", alpha=1, c=color, s=100)
plt.savefig("../Plots/Salary versus FT%.jpg")
plt.show()
```

```python
player_normalized['eFG%_bin'] = player_normalized['eFG%']//0.05*0.05
player_normalized.head()
player_gby_efg = player_normalized.groupby("eFG%_bin")
player_gby_efg.describe()
player_mean_efg_df = player_gby_efg.mean()
player_mean_efg_df = player_mean_efg_df[player_mean_efg_df.index < 1 ]
player_mean_efg_df
```

```python
x = player_mean_efg_df.index
y = player_mean_efg_df["Salary Normalized (Million)"]
color = ["steelblue"]
plt.xlabel('Effective Field Goal Percentage')
plt.ylabel('Average Salary (Million) Normalized to 2018')
plt.title("Salary versus eFG%")
plt.scatter(x, y, marker=".", alpha=1, c=color, s=100)
plt.savefig("../Plots/Salary versus eFG%.jpg")
plt.show()
```
